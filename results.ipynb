```python
import matplotlib.pyplot as plt
import numpy as np

models = ['Llama 3.2', 'Qwen 2.5', 'BERT']
before_finetune = [0.42, 0.45, 0.26]
after_finetune = [0.45, 0.53, 0.34]

bar_width = 0.3
index = np.arange(len(models))

fig, ax = plt.subplots(figsize=(8, 5))

bar1 = ax.bar(index, before_finetune, bar_width, label='Before Fine-tuning', color='lightcoral')
bar2 = ax.bar(index + bar_width, after_finetune, bar_width, label='After Fine-tuning', color='seagreen')
ax.set_xlabel('Model')
ax.set_ylabel('Accuracy')
ax.set_title('Comparison of Accuracy Before and After Fine-tuning')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(models)
ax.legend()

# Set y-axis limits
ax.set_ylim(0, 0.6)

for bar in bar1:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 0.01, f'{yval:.2f}', ha='center', va='bottom')

for bar in bar2:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 0.01, f'{yval:.2f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
